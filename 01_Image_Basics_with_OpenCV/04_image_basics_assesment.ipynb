```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img=cv2.imread("../DATA/dog_backpack.jpg")
```

```python
plt.imshow(img)
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
fimg_rgb=cv2.flip(img_rgb,0)
plt.imshow(fimg_rgb)
```

```python
cv2.rectangle(img_rgb,(200,400),(600,700),(255,0,0),thickness=5)
plt.imshow(img_rgb)
```

```python
p1=(400,400)
p2=(200,700)
p3=(600,700)
cv2.line(img_rgb,p1,p2,(0,0,255),thickness=5)
cv2.line(img_rgb,p2,p3,(0,0,255),thickness=5)
cv2.line(img_rgb,p3,p1,(0,0,255),thickness=5)
plt.imshow(img_rgb)
```
