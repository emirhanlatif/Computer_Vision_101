```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
blank_img=np.zeros((512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
blank_img
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,(100,200),(300,300),(255,0,0),thickness=3)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,(100,100),radius=50,color=(0,255,0),thickness=-1)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,(150,200),(300,200),(0,0,255),thickness=5)
plt.imshow(blank_img)
```

```python
cv2.putText(blank_img,"Hello",org=(100,250),fontFace=cv2.FONT_HERSHEY_COMPLEX,fontScale=3,color=(0,200,0),thickness=3)
plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[100,300],[200,200],[400,300],[200,400]],np.int32)
```

```python
pts = vertices.reshape((-1,1,2))  # çokgen çizmek için bu yapıyı kullanmalıyız
```

```python
vertices
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
