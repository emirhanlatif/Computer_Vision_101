```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def display_img(img,cmap=None):
    fig=plt.figure(figsize=(12,10))
    ax=fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg')
giaraffes_rgb=cv2.cvtColor(giaraffes,cv2.COLOR_BGR2RGB)
display_img(giaraffes_rgb)
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg',0)
ret,thresh=cv2.threshold(giaraffes,127,255,cv2.THRESH_BINARY)
```

```python
display_img(thresh,cmap='gray')
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg')
giaraffes_hsv=cv2.cvtColor(giaraffes,cv2.COLOR_BGR2HSV)
display_img(giaraffes_hsv)
```

```python
kernel =np.ones((4,4),dtype=np.float32)/10
```

```python
kernel
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg')
giaraffes_rgb=cv2.cvtColor(giaraffes,cv2.COLOR_BGR2RGB)
dst = cv2.filter2D(giaraffes_rgb,-1,kernel)
display_img(dst)
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg',0)
sobelx=cv2.Sobel(giaraffes,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx,cmap='gray')
```

```python
giaraffes=cv2.imread('../DATA/giraffes.jpg')
img=giaraffes
color=('b','g','r')
for i,col in enumerate(color):
    histr=cv2.calcHist([img],channels=[i],mask=None,histSize=[256],ranges=[0,256])
    plt.plot(histr,color=col)
    plt.xlim([0,256])
plt.show()
```
